```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor


mse_sum_dt = 0
mse_sum_rf = 0
mse_sum_gbr = 0
for i in range(500):
    x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 0)
    dt_val = DecisionTreeRegressor()
    dt_val.fit(x_train,y_train)
    y_val_pre_dt = dt_val.predict(x_val)
    mse_sum_dt += mean_squared_error(y_val,y_val_pre_dt)
    
    rf_val = RandomForestRegressor()
    rf_val.fit(x_train,y_train)
    y_val_pre_rf = rf_val.predict(x_val)
    mse_sum_rf += mean_squared_error(y_val,y_val_pre_rf)
    
    gbr_val = GradientBoostingRegressor()
    gbr_val.fit(x_train,y_train)
    y_val_pre_gbr = gbr_val.predict(x_val)
    mse_sum_gbr += mean_squared_error(y_val,y_val_pre_gbr)
mse_dt = mse_sum_dt/500
mse_rf = mse_sum_rf/500
mse_gbr = mse_sum_gbr/500
```
