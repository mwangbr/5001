```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("trainprocess_2.csv")
test_data = pd.read_csv("testprocess_2.csv")
```

```python
X_train = train_data.drop(columns = ["id","playtime_forever","genres","categories","tags","purchase_date","release_date"])
Y_train = train_data["playtime_forever"]
X_test = test_data.drop(columns = ["id","genres","categories","tags","purchase_date","release_date"])
```

```python
X_train.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train,Y_train)
```

```python
feat_labels = X_train.columns[:]
importances = dt.feature_importances_
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
```

```python
new_train = X_train[["Difficult","total_positive_reviews","Shared/Split Screen","Fantasy","2D","consider_Year","release_Month",\
                    "total_reviews","release_Day","total_negative_reviews","Military","Alternate History","Open World"]]#,"Survival",\
                    #"In-App Purchases","Dungeon Crawler","Cartoony","Nudity","total_positive_reviews_rate","Action-Adventure",\
                     #"Exploration","Based On A Novel","purchase_Day"]]
new_test = X_test[["Difficult","total_positive_reviews","Shared/Split Screen","Fantasy","2D","consider_Year","release_Month",\
                    "total_reviews","release_Day","total_negative_reviews","Military","Alternate History","Open World"]]#,"Survival",\
                    #"In-App Purchases","Dungeon Crawler","Cartoony","Nudity","total_positive_reviews_rate","Action-Adventure",\
                     #"Exploration","Based On A Novel","purchase_Day"]]
```

```python
mse_sum_dt = 0
mse_sum_rf = 0
mse_sum_gbr = 0
for i in range(500):
    x_train, x_val, y_train, y_val = train_test_split(new_train, Y_train, test_size = 0.3, random_state = 0)
    dt_val = DecisionTreeRegressor()
    dt_val.fit(x_train,y_train)
    y_val_pre_dt = dt_val.predict(x_val)
    mse_sum_dt += mean_squared_error(y_val,y_val_pre_dt)
    
    rf_val = RandomForestRegressor()
    rf_val.fit(x_train,y_train)
    y_val_pre_rf = rf_val.predict(x_val)
    mse_sum_rf += mean_squared_error(y_val,y_val_pre_rf)
    
    gbr_val = GradientBoostingRegressor()
    gbr_val.fit(x_train,y_train)
    y_val_pre_gbr = gbr_val.predict(x_val)
    mse_sum_gbr += mean_squared_error(y_val,y_val_pre_gbr)
mse_dt = mse_sum_dt/500
mse_rf = mse_sum_rf/500
mse_gbr = mse_sum_gbr/500
```

```python
print(mse_dt)
print(mse_rf)
print(mse_gbr)
```
