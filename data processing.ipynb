```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("train.csv")
```

```python
train_data.dropna(axis = 0,inplace = True)
train_data.info()
train_data.head()
```

```python
train_data["total_reviews"] =train_data["total_positive_reviews"]+train_data["total_negative_reviews"]
train_data["total_positive_reviews_rate"] = train_data["total_positive_reviews"]/ train_data["total_reviews"]
train_data["total_negative_reviews_rate"] = train_data["total_negative_reviews"]/ train_data["total_reviews"]
list1 = list(train_data["total_positive_reviews_rate"])
list2 = list(train_data["total_negative_reviews_rate"])
total_positive_reviews_rate_list = []
total_negative_reviews_rate_list = []
for rate in list1:
    if np.isnan(rate) == True:
        total_positive_reviews_rate_list.append(0)
    else:
        total_positive_reviews_rate_list.append(rate)
for rate in list2:
    if np.isnan(rate) == True:
        total_negative_reviews_rate_list.append(0)
    else:
        total_negative_reviews_rate_list.append(rate)
train_data["total_positive_reviews_rate"] = total_positive_reviews_rate_list
train_data["total_negative_reviews_rate"] = total_negative_reviews_rate_list
```

```python
train_data["purchase_date"] = pd.to_datetime(train_data["purchase_date"])
train_data["release_date"] = pd.to_datetime(train_data["release_date"])

train_data["purchase_Year"] = train_data["purchase_date"].dt.year
train_data["purchase_Month"] = train_data["purchase_date"].dt.month
train_data["purchase_Day"] = train_data["purchase_date"].dt.day
train_data["purchase_Weekday"] = train_data["purchase_date"].dt.dayofweek

train_data["release_Year"] = train_data["release_date"].dt.year
train_data["release_Month"] = train_data["release_date"].dt.month
train_data["release_Day"] = train_data["release_date"].dt.day
train_data["release_Weekday"] = train_data["release_date"].dt.dayofweek
```

```python
train_data["consider_Year"] = train_data["purchase_Year"]-train_data["release_Year"]
```

```python
print(train_data["consider_Year"])
print(train_data["purchase_Year"])
print(train_data["release_Year"])
```

```python
genres = train_data["genres"].to_list()
genres_list = []
new_genres_list = []
for genre in genres:
    genres_list.append(genre.split(","))
for genre in genres_list:
    new_list = []
    for gen in genre:
        new_list.append("genres_" + gen)
    new_genres_list.append(",".join(new_list))
train_data["genres"] = new_genres_list
train_data
```

```python
onehot_genres = train_data["genres"].str.get_dummies(sep = ",")
onehot_categories = train_data["categories"].str.get_dummies(sep = ",")
onehot_tags = train_data["tags"].str.get_dummies(sep = ",")
```

```python
train_data = pd.concat([train_data, onehot_genres,onehot_categories,onehot_tags], axis=1)
```

```python
test_data = pd.read_csv("test.csv")
test_data.info()
```

```python
test_genres = test_data["genres"].to_list()
test_genres_list = []
test_new_genres_list = []
for genre in test_genres:
    test_genres_list.append(genre.split(","))
test_genres_list
for genre in test_genres_list:
    test_new_list = []
    for gen in genre:
        test_new_list.append("genres_" + gen)
    test_new_genres_list.append(",".join(test_new_list))
test_data["genres"] = test_new_genres_list
test_data
```

```python
test_data["purchase_date"] = pd.to_datetime(test_data["purchase_date"])
test_data["release_date"] = pd.to_datetime(test_data["release_date"])

test_data["purchase_Year"] = test_data["purchase_date"].dt.year
test_data["purchase_Month"] = test_data["purchase_date"].dt.month
test_data["purchase_Day"] = test_data["purchase_date"].dt.day
test_data["purchase_Weekday"] = test_data["purchase_date"].dt.dayofweek

test_data["release_Year"] = test_data["release_date"].dt.year
test_data["release_Month"] = test_data["release_date"].dt.month
test_data["release_Day"] = test_data["release_date"].dt.day
test_data["release_Weekday"] = test_data["release_date"].dt.dayofweek
```

```python
test_data["purchase_Year"].fillna(test_data["purchase_Year"].mean(), inplace=True)
test_data["purchase_Month"].fillna(test_data["purchase_Month"].mean(), inplace=True)
test_data["purchase_Day"].fillna(test_data["purchase_Day"].mean(), inplace=True)
test_data["purchase_Weekday"].fillna(test_data["purchase_Weekday"].mean(), inplace=True)
test_data.info()
```

```python
test_data["consider_Year"] = test_data["purchase_Year"]-test_data["release_Year"]
```

```python
test_data["total_positive_reviews"].fillna(test_data["total_positive_reviews"].mean(),inplace = True)
test_data["total_negative_reviews"].fillna(test_data["total_negative_reviews"].mean(),inplace = True)
test_data["total_reviews"] =test_data["total_positive_reviews"]+test_data["total_negative_reviews"]

test_data["total_positive_reviews_rate"] = test_data["total_positive_reviews"]/ test_data["total_reviews"]
test_data["total_negative_reviews_rate"] = test_data["total_negative_reviews"]/ test_data["total_reviews"]
list_p = list(test_data["total_positive_reviews_rate"])
list_n = list(test_data["total_negative_reviews_rate"])
test_total_positive_reviews_rate_list = []
test_total_negative_reviews_rate_list = []
for rate in list_p:
    if np.isnan(rate) == True:
        test_total_positive_reviews_rate_list.append(0)
    else:
        test_total_positive_reviews_rate_list.append(rate)
for rate in list_n:
    if np.isnan(rate) == True:
        test_total_negative_reviews_rate_list.append(0)
    else:
        test_total_negative_reviews_rate_list.append(rate)
test_data["total_positive_reviews_rate"] = test_total_positive_reviews_rate_list
test_data["total_negative_reviews_rate"] = test_total_negative_reviews_rate_list
```

```python
test_onehot_genres = test_data["genres"].str.get_dummies(sep = ",")
test_onehot_categories = test_data["categories"].str.get_dummies(sep = ",")
test_onehot_tags = test_data["tags"].str.get_dummies(sep = ",")
test_data = pd.concat([test_data, test_onehot_genres,test_onehot_categories,test_onehot_tags], axis=1)
```
